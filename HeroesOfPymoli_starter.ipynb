## Gender Demographics

* Percentage and Count of Male Players


* Percentage and Count of Female Players


* Percentage and Count of Other / Non-Disclosed




```python
#gender stufff
print(purchase_data)
male_count = purchase_data["Gender"].value_counts()['Male']
female_count = purchase_data["Gender"].value_counts()['Female']
na_count = purchase_data["Gender"].value_counts()['Other / Non-Disclosed']
total_gender = male_count + female_count + na_count

percentage_male = male_count / total_gender * 100
percentage_male.round(2)
print(percentage_male)

percentage_female = female_count / total_gender * 100
percentage_female.round(2)
print(percentage_female)

percentage_na = na_count / total_gender * 100
percentage_na.round(2)
print(percentage_na)
purchase_data()
```


## Purchasing Analysis (Gender)

* Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. by gender




* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame

```python
#purchase analysis gender

#purchase count
total_purchase = purchase_data.groupby("Gender")["Item Name"].value_counts()
total_price = purchase_data.groupby("Gender")["Price"].sum()
mean_spent = purchase_data.groupby("Gender")["Price"].mean()

#yuh average spent
average_spent = (total_price / total_purchase)


gender_dem = pd.DataFrame({"Purchase Count":total_purchase, 
                                   "Average Purchase Price":average_spent,
                                   "Total Purchases":total_price,})
gender_dem.round(2)
```

## Age Demographics

* Establish bins for ages


* Categorize the existing players using the age bins. Hint: use pd.cut()


* Calculate the numbers and percentages by age group


* Create a summary data frame to hold the results


* Optional: round the percentage column to two decimal points


* Display Age Demographics Table


```python
#bins for ages
bins = [0, 10, 18, 30, 45, 55, 75]
age_groups = ["Baby","Teen", "Alchohol", "Medium Old", "Fuckin Old", "Deathbed"]
print(age_groups)
purchase_data["Age Group"] = pd.cut(purchase_data["Age"], bins, labels=age_groups)
purchase_data.groupby("Age Group")

age_average = purchase_data.groupby("Age Group")["Price"].mean()
age_average.round(2)
age_groups_spending = purchase_data.groupby("Age Group")["Item Name"]
age_groups_spending.count()
age_group_purchase = purchase_data.groupby("Age Group")["Price"].sum()
age_total= age_group_purchase / 780
age_total.round(2)

age_group_done = pd.DataFrame({"Purchase Count":age_groups_spending,
                            "Average Purchase Price":age_average,
                            "Total Purchases":age_group_purchase,
                            "Totals":age_total})

age_group_done.head()
```
